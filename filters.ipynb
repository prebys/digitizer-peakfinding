```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, filtfilt

# Generate some sample data
t = np.linspace(0, 1, 1000)  # 1 second time interval
signal = np.sin(2 * np.pi * 5 * t)  # 5 Hz sine wave
noise = 0.2 * np.random.randn(1000)  # Gaussian noise
noisy_signal = signal + noise

# Wrapper for Butterworth filter
def filter(data, cutoff_freq, fs, btype='low',order=5):
    nyquist = 0.5 * fs
    normal_cutoff = cutoff_freq / nyquist
    b, a = butter(order, normal_cutoff, btype=btype, analog=False)
    filtered_data = filtfilt(b, a, data)
    return filtered_data
```

```python
# Set filter parameters
cutoff_freq = 10  # cutoff frequency in Hz
fs = 1000  # sampling frequency in Hz
order = 6  # filter order

# Apply filter
filter_btype = 'high'
filtered_signal = filter(noisy_signal, cutoff_freq, fs, filter_btype, order)

# Plot the original and filtered signals
plt.figure(figsize=(10, 6))
plt.plot(t, noisy_signal, label='Noisy Signal')
plt.plot(t, filtered_signal, label='Filtered Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title(filter_btype+'-pass Filter Example')
plt.legend()
plt.grid(True)
plt.show()

# Apply filter
filter_btype = 'low'
filtered_signal = filter(noisy_signal, cutoff_freq, fs, filter_btype, order)

# Plot the original and filtered signals
plt.figure(figsize=(10, 6))
plt.plot(t, noisy_signal, label='Noisy Signal')
plt.plot(t, filtered_signal, label='Filtered Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title(filter_btype+'-pass Filter Example')
plt.legend()
plt.grid(True)
plt.show()
```
